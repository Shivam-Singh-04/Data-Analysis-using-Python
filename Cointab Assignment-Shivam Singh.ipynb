```python
#Let's find totel weight in (kg). Here, we will merge sku_master and order_report on the basis of SKU column and then applying group by on the basis of order number forming a new dataframe as total_weight.
total_weight = sku_master.merge(order_report,how = 'inner',on = 'SKU')
total_weight['Total Weight (Kg)'] = (total_weight['Weight (g)']*total_weight['Order Qty'])/1000
total_weight.drop_duplicates()

total_weight = total_weight.groupby('ExternOrderNo',as_index = False ).sum('Total Weight (Kg)')[['ExternOrderNo','Total Weight (Kg)']]
```

```python
# Getting overview of total weights
total_weight.head(10)
```

```python
# Merging total_weight with invoice_data on Extern Order No from total_weight and Order ID from invoice_data to compare the weights in both the courier invoice and as per company.
temp_invoice = invoice_data.merge(total_weight,how = 'inner',left_on = 'Order ID',right_on = 'ExternOrderNo')
```

```python
# Checking contents of temp_invoice
temp_invoice.head(10)
```

```python
# Now, we will concatenate temp_invoice with pincode_zones by company to compare the zones if they are applied correctly and renaming the zones column in the pincode_zones dataframe as 'Zone by company' to distinguish.
final= pd.concat([temp_invoice,pincode_zones],axis = 1)
final.columns.values[12] = 'Zone_by_company'
```

```python
# Checking contents of final dataframe
final
```

```python
# Let's define a ceiling value to calculate and round off the shipping charges both by company and the courier service provider
np.ceil(0.515/0.5)*0.5/0.5-1
```

```python
# Calculating weight slab as per courier company in a new column weight_slab
def slab_calculate():
  slab = []
  for i in final.index:
    d = final.iloc[i]
    if d['Zone'] == 'd':
     slab.append(np.ceil(d['Charged Weight']/1.25)*1.25)
    elif d['Zone'] == 'a':
      slab.append(np.ceil(d['Charged Weight']/.25)*.25)
    elif d['Zone'] == 'b':
      slab.append(np.ceil(d['Charged Weight']/.50)*.50)
    elif d['Zone'] == 'c':
      slab.append(np.ceil(d['Charged Weight']/.75)*.75)
    elif d['Zone'] == 'e':
      slab.append(np.ceil(d['Charged Weight']/1.00)*1.00)
  return slab


final.insert(loc = 3,
          column = 'weight_slab',
          value = slab_calculate())
```

```python
# Calculating weight slab as per company X in a new column weight_slab_by_X
def slab_calculater_X():
  slab = []
  for i in final.index:
    d = final.iloc[i]
    if d['Zone_by_company'] == 'd':
      slab.append(np.ceil(d['Total Weight (Kg)']/1.25)*1.25)
    elif d['Zone_by_company'] == 'a':
      slab.append(np.ceil(d['Total Weight (Kg)']/.25)*.25)
    elif d['Zone_by_company'] == 'b':
      slab.append(np.ceil(d['Total Weight (Kg)']/.50)*.50)
    elif d['Zone_by_company'] == 'c':
      slab.append(np.ceil(d['Total Weight (Kg)']/.75)*.75)
    elif d['Zone_by_company'] == 'e':
      slab.append(np.ceil(d['Total Weight (Kg)']/1.00)*1.00)
  return slab

final.insert(loc = 10,
          column = 'weight_slab_by_X',
          value = slab_calculater_X())
```

```python
# Checking above two newly added columns in final dataframe
final
```

```python
# Calaulating the total delivery charges as per the slab by company X and adding a new column charges_as_per_X. Since only forward and forward & RTO charges have been made, we will include both these charges only.
def charges():
    charge = []
    for i in final.index:
        d =  final.iloc[i]
        if d['Zone_by_company'] =='a':
            if d['Type of Shipment'] == 'Forward charges':
                fixed_charge = 29.5
                additional_charge = 23.6*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
            elif d['Type of Shipment'] == 'Forward and RTO charges':
                fixed_charge = 29.5+13.6
                additional_charge = 2*23.6*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
        elif d['Zone_by_company'] =='b':
            if d['Type of Shipment'] == 'Forward charges':
                fixed_charge = 33
                additional_charge = 28.3*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
            elif d['Type of Shipment'] == 'Forward and RTO charges':
                fixed_charge = 33+20.5
                additional_charge = 2*28.3*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
        elif d['Zone_by_company'] =='c':
            if d['Type of Shipment'] == 'Forward charges':
                fixed_charge = 40.1
                additional_charge = 38.9*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
            elif d['Type of Shipment'] == 'Forward and RTO charges':
                fixed_charge = 40.1+31.9
                additional_charge = 2*38.9*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
        elif d['Zone_by_company'] =='d':
            if d['Type of Shipment'] == 'Forward charges':
                fixed_charge = 45.4
                additional_charge = 44.8*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
            elif d['Type of Shipment'] == 'Forward and RTO charges':
                fixed_charge = 45.4+41.3
                additional_charge = 2*44.8*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
        elif d['Zone_by_company'] =='e':
            if d['Type of Shipment'] == 'Forward charges':
                fixed_charge = 56.6
                additional_charge = 55.5*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
            elif d['Type of Shipment'] == 'Forward and RTO charges':
                fixed_charge = 56.6+50.7
                additional_charge = 2*55.5*(d['weight_slab_by_X']/1.25-1)
                total_charge = fixed_charge+additional_charge
                charge.append(total_charge)
    return charge
    

final.insert(loc = 15,
          column = 'charges_as_per_X',
          value = charges())
```

```python
# checking final dataframe before calculating thr difference between the shipping charges for company X and courier companies.
final
```

```python
# Now, we will calculate the difference between the shipping charges for company X and courier companies in a new column named as 'difference_b/w_charges'.
final['difference_b/w_charges'] =round((final['charges_as_per_X'] -final['Billing Amount (Rs.)']) ,2)
```

```python
# The final dataframe indicating difference between the shipping charges for company X and courier companies.
# Here negative value means the courier company have been paid more than the actual charges, 0 means no extra charges and positive means the courier company is underpaid. 
final
```
